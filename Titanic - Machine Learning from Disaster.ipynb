```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic=pd.read_csv('train.csv')
```

```python
titanic
```

```python
titanic.shape
```

```python
titanic.describe()
```

```python
#dummies- way to represent categorical data
ports=pd.get_dummies(titanic.Embarked,prefix="Embarked")
ports.head()
```

```python
#joining port to original datset
titanic=titanic.join(ports)
titanic
```

```python
titanic.drop(['Embarked'],axis=1,inplace=True)
titanic
```

```python
titanic.Sex=titanic.Sex.map({'male':0,'female':1})
titanic
```

```python
y=titanic.Survived.copy()
x=titanic.drop(['Survived'],axis=1)
```

```python
print(x.shape)
print(y.shape)
```

```python
x.drop(['Cabin','Ticket','Name','PassengerId'],axis=1,inplace=True)
x
```

```python
x.info()
```

```python
x.isnull().values.any()  #places where there is null
```

```python
x[pd.isnull(x).any(axis=1)] #searching and display null values
```

```python
x.Age.fillna(x.Age.mean(),inplace=True)
```

```python
x.isnull().values.any() 
```

```python
from sklearn.model_selection import train_test_split #random state for randomness in data every time you run
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model.score(x_test,y_test)
```
